```python
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

%matplotlib inline

# virtual buffer size
# vbs = (  '100', '80', ' 60',  '40',  '20',  '10', '16',   '32',   '48', 'EPB-16V','NR', 'NRW'  ,'DB'   ,'CHN', 'InOS' )
# t_vbs = [0.297, 0.277, 0.282, 0.274, 0.276,  0.28, 0.196, 0.1997, 0.276, 0.08,    0.07 , 0.046 , 0.047 ,0.054, 0.052]

vbs = ( ' 60', '48',     '40',  '32',    '20',  '16',  '10',    'EPB/16' , 'NR', 'NRW' )
t_vbs = [0.282, 0.276,    0.274, 0.1997,  0.276,  0.196, 0.28,   0.08 , 0.07 , 0.046 ]


f = plt.figure()
pos_list = np.arange(len(t_vbs))
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list)))
ax.xaxis.set_major_formatter(ticker.FixedFormatter((vbs)))
plt.ylabel('Seconds')
plt.xlabel('Virtualized buffer size')
plt.bar(pos_list, t_vbs, align = 'center', color='0.5', hatch='/')
plt.tight_layout()
f.savefig("virtual_buffer_size.eps",format="eps", dpi=1200)
plt.show()
```

```python
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

%matplotlib inline

# virtual buffer size
vbs = ['IPOS(EPB-16V)' ,'IPOS(DB)', 'CHN', 'InOS'  ]

t_vbs_ipos = [0.08 , 0.047 ]
ipos_list = np.arange(len(t_vbs_ipos))

t_vbs_chain = [0.054]
chain_list = np.arange(len(t_vbs_chain)) + len(t_vbs_ipos)

t_vbs_InOS = [0.052]
InOS_list = np.arange(len(t_vbs_InOS)) + len(t_vbs_ipos) + len(t_vbs_chain)


f = plt.figure()

ax = plt.axes()

ax.xaxis.set_major_locator(ticker.FixedLocator(( np.arange(len(vbs)) ) ) )
ax.xaxis.set_major_formatter(ticker.FixedFormatter((vbs)))
plt.ylabel('Seconds')
#plt.xlabel('')

plt.bar(ipos_list,  t_vbs_ipos, align = 'center', hatch = '/', color= '0.5' )
plt.bar(chain_list, t_vbs_chain,align = 'center', hatch = '+', color = '0.8')
plt.bar(InOS_list,  t_vbs_InOS, align = 'center', hatch = 'x', color = '0.8')

plt.tight_layout()
f.savefig("IPOS_Chain_InOS.eps",format="eps", dpi=1200)
plt.show()
```

```python
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

%matplotlib inline

# virtual buffer size
vbs = ( '1 RT', '2 RT', '4 RT'  , '6 RT'   ,'12 RT', '24 RT', 'CHN'  )
t_vbs_ipos = [15.62, 10.01 ,  7.23 ,   6.281    ,5.367, 4.88]
ipos_list = np.arange(len(t_vbs_ipos))

t_vbs_chain = [4.418]
chain_list = np.arange(len(t_vbs_chain)) + len(t_vbs_ipos)


f = plt.figure()
ax = plt.axes()


ax.xaxis.set_major_locator(ticker.FixedLocator(( np.arange(len(vbs)) ) ) )
ax.xaxis.set_major_formatter(ticker.FixedFormatter((vbs)))
plt.ylabel('milliseconds')
plt.xlabel('Virtual Task size')

plt.bar(ipos_list,  t_vbs_ipos, align = 'center', hatch = '/', color= '0.5' )
plt.bar(chain_list, t_vbs_chain,align = 'center', hatch = '+', color = '0.8')

plt.tight_layout()
f.savefig("virtualTaskSize.eps",format="eps", dpi=1200)
plt.show()
```
